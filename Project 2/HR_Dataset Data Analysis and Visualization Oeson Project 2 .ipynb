**Question 4: Performance based on Department: Which department has the best average performance (please note that 'PerfScoreID' is the categorical representation of 'PerformanceScore')**

```python
# A. Average Performance by Department
sorted_perf = hr_data.groupby('Department')['PerfScoreID'].mean().sort_values(ascending=False).reset_index()

plt.figure(figsize=(8, 6))
sns.barplot(x='Department', y='PerfScoreID', data=sorted_perf, palette='Set1')

plt.title('Average Performance by Department', fontsize=15, weight='bold')
plt.xlabel('Department', fontsize=10, weight='bold')
plt.xticks(rotation=45)
plt.ylabel('Average Performance Score', fontsize=10, weight='bold')
plt.show()
```

```python
#### B. Breakdown of Performance in each Department

department_performance_data = hr_data.groupby(['Department', 'PerfScoreID']).size().unstack(fill_value=0)

# Plotting pie charts for each department
n_rows = 2
n_cols = 3
fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, 10))  # Adjust figsize as needed
axes = axes.flatten()  # Flatten the array of axes for easy iteration

# Define colors for each PerfScoreID, assuming they are 1, 2, 3, 4
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']  # Replace with your colors

# Match the Performance score Id and Performance score
performance_score_mapping=dict()
for  id,score in zip(hr_data['PerfScoreID'].unique(),hr_data['PerformanceScore'].unique()):
    performance_score_mapping[id]=score

for ax, (department, scores) in zip(axes, department_performance_data.iterrows()):
    # Only include labels for segments that are not 0%
    labels = [performance_score_mapping[score] if size > 0 else '' for score, size in scores.items()]
    wedges, texts, autotexts = ax.pie(scores, labels=labels, autopct=lambda p: '{:.1f}%'.format(p) if p > 2 else '', startangle=90)
    ax.set_title(department, fontweight='bold', fontsize=14)

    # Increase the size of the autopct text
    for autotext in autotexts:
        autotext.set_size(10)

# Remove the empty subplots (if the number of departments is less than 6)
for ax in axes[len(department_performance_data):]:
    ax.remove()
    
# Creating a legend using a dummy axis
handles, labels = ax.get_legend_handles_labels()
plt.subplots_adjust(right=0.8)  
plt.suptitle('Performance Breakdown in Each Department', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()
```

**Answer 4: Performance varies by department: IT/IS and Software Engineering show diverse results, from top performers to those on improvement plans, while Admin Offices consistently meet expectations. The bar plot reveals differences in average scores, with Engineering and Sales typically scoring higher, suggesting department-specific influences.**

**Question 10: Recruitment Source and Compensation Comparison: How does the recruitment source influence salary levels for different positions?**

```python

# Create the pivot table for average salary
salary_pivot = hr_data.pivot_table('Salary', index='Position', columns='RecruitmentSource', aggfunc='mean')

# Generate the ranking table
salary_rankings = salary_pivot.rank(axis=1, method='min', ascending=False)

# Function to combine salary and rank
def combine_salary_rank_simple(salary, rank):
    if pd.isna(salary):
        return "NaN"
    else:
        return f"{salary:.0f} ({int(rank)})"

# Creating a new DataFrame for the combined data
combined_data = pd.DataFrame(index=salary_pivot.index, columns=salary_pivot.columns)

# Filling the new DataFrame with combined salary and rank data
for col in salary_pivot.columns:
    for idx in salary_pivot.index:
        salary = salary_pivot.at[idx, col]
        rank = salary_rankings.at[idx, col]
        combined_data.at[idx, col] = combine_salary_rank_simple(salary, rank)

# Plotting the heatmap
plt.figure(figsize=(15, 10))
sns.heatmap(combined_data.applymap(lambda x: None if x == "NaN" else float(x.split()[0])), 
            annot=combined_data, fmt="", cmap="YlGnBu", linewidths=.5)
plt.title('Heatmap of Average Salary by Position and Recruitment Source (with Rankings)', fontsize=15, weight='bold')
plt.xlabel('Recruitment Source', fontsize=15, weight='bold')
plt.ylabel('Position', fontsize=15, weight='bold')
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.show()
```

**Answer 10: Job seekers should note that candidates from the Diversity Job Fair tend to secure higher-paying roles across various positions. Strategic application through proven sources can influence starting salaries.**
